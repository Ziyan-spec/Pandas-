# **Merging two DataFrame** **bold text**

```python
# Employees table
employees = pd.DataFrame({
    "EmployeeID": ["E101", "E102", "E103", "E104", "E106"],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Department": ["HR", "IT", "Finance", "IT", "Development"],
    "JoiningYear": [2019, 2020, 2021, 2025, 2022]
})

# Salaries table
salaries = pd.DataFrame({
    "EmployeeID": ["E101", "E102", "E105", "E104", "E107"],
    "Salary": [50000, 90000, 45000, 70000, 40000],
    "Bonus": [5000, 6000, 4000, 2500, 3000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='EmployeeID')
```

```python
pd.merge(employees, salaries, on='EmployeeID', how='inner')
```

```python
pd.merge(employees, salaries, on='EmployeeID', how='outer')
```

```python
pd.merge(employees, salaries, on='EmployeeID', how='left')
```

```python
pd.merge(employees, salaries, on='EmployeeID', how='right')
```

# **GroupBy**

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S1', 'S2', 'S1', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
df
```

```python
#groupby category and find sum of sales
df.groupby('Category')['Sales'].sum()
```

```python
#groupby store and find sum of sales
a = df.groupby('Store')['Sales']
for i,j in a:
    print(i)
    print(j)
```

```python
a.sum()
```

```python
#groupby on multiple columns
df.groupby(['Category','Store'])['Sales'].sum()
```

# **Pivot tables**

```python
# sample dataset for pivot tables
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South'] * 5,
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice',
            'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice']
 }
df = pd.DataFrame(data)
df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)

df
```

```python
p1 = pd.pivot_table(df, values='Sales', index='Region', columns='Product', aggfunc='median')
p1
```

```python
p2 = pd.pivot_table(df, values=['Sales','Units'], index='Region', columns='Product', aggfunc='mean')
p2
```

```python
#crosstab is used for counting purpose
pd.crosstab(df['Product'], df['Region'])
```
